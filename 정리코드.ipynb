```python
# module - init.py
# 비어있음

# dataloader.py
import pandas as pd

def get_etf_info(etf_dict):
    etf_info_df = etf_dict['general_info']
    etf_type = etf_info_df[['Name','유형분류(대),유형분류(중), 유형분류(소)']]
    
    # 국내주식
    kor_stock_etf = etf_type[etf_type['유형분류(대)']=='국내주식형']
    kor_stock_etf['대신분류'] = '국내주식'
    
    # 국내채권
    kor_bond_etf = etf_type[etf_type['유형분류(대)']=='국내채권형']
    kor_bond_etf['대신분류'] = '국내채권'
    
    
    # 선진국주식
    developed_stock_etf_1 = etf_type[(etf_type['유형분류(대)']=='해외주식형')&
                                    (etf_type['유형분류(중)']=='권역별')&
                                     (etf_type['유형분류(소)']=='북미주식')|(etf_type['유형분류(소)']=='유롭주식')|(etf_type['유형분류(소)']=='글로벌주식')
                                    ]
    
    developed_stock_etf_2 =  etf_type[(etf_type['유형분류(대)']=='해외주식형')&
                                    (etf_type['유형분류(중)']=='국가별')&
                                     (etf_type['유형분류(소)']=='일본주식')]
    
    
    developed_stock_etf = pd.concat([developed_stock_etf_1,developed_stock_etf_2],axis=0)
    developed_stock_etf['대신분류'] = '선진국주식'
    
    
    # 이머징 주식
    developing_stock_etf_1 = etf_type[(etf_type['유형분류(대)']=='해외주식형')&
                               (etf_type['유형분류(중)']=='권역별')&
                                (etf_type['유형분류(소)']=='신흥국주식')|(etf_type['유형분류(소)']=='아시아퍼시픽주식')|(etf_type['유형분류(소)']=='중남미주식')
                               ]
    developing_stock_etf_2 =  etf_type[(etf_type['유형분류(대)']=='해외주식형')&
                               (etf_type['유형분류(중)']=='국가별')&
                                (etf_type['유형분류(소)']=='중국주식')|(etf_type['유형분류(소)']=='베트남주식')|(etf_type['유형분류(소)']=='인도주식')|(etf_type['유형분류(소)']=='러시아주식')|(etf_type['유형분류(소)']=='기타국가주식')]
    
    developing_stock_etf = pd.concat([developing_stock_etf_1,developing_stock_etf_2],axis=0)
    
    developing_stock_etf['대신분류'] = '이머징주식'

    
    # 선진국 채권
    developed_bond_etf = etf_type[(etf_type['유형분류(대)']=='해외채권형')&
                                    (etf_type['유형분류(중)']=='지역채권')&
                                     (etf_type['유형분류(소)']=='북미채권')]
    
    developed_bond_etf['대신분류'] = '선진국주식'

    
    # 하이일드 채권
    # 원자재
    raw_material_df = etf_type[(etf_type['유형분류(대)']=='해외대체')&(etf_type['유형분류(중)']=='해외특별자산')&(etf_type['유형분류(소)']=='해외원자재')]
    
    raw_material_df['대신분류'] = '원자재'
    
    
    # 외화
    # 달러
    foreign_exchange_eft = eft_type.loc[['A261240','A138230']]
    
    foreign_exchange_eft['대신분류'] = '달러'
    
    # 부동산/리츠
    realty_etf_1 =  etf_type[(etf_type['유형분류(대)']=='해외대체')&(etf_type['유형분류(중)']=='해외부동산')]
    realty_etf_2 =  etf_type[(etf_type['유형분류(대)']=='rnrso대체')&(etf_type['유형분류(중)']=='부동산')]
    
    realty_etf = pd.concat([realty_etf_1,realty_etf_2], axis=0)
    realty_etf['대신분류'] = '부동산'
    
    tot_etf_df = pd.concat([kor_stock_etf, kor_bond_etf, developed_stock_etf,developing_stock_etf,developed_bond_etf,
                           raw_material_df,foreign_exchange_eft,realty_etf], axis=0)
    
    tot_etf_df= tot_etf_df[~ tot_etf_df.Name.str.contains("레버리지")]
        
    return tot_etf_df
```

```python
# expected_returns.py

import warnings
import pandas as pd
import numpy as np


def returns_from_prices(prices, log_returns=False):
    """
    calculate the returns given prices.
    
    :param prices: adjusted (daily) closing prices of the asset, each row is a date and each column is a ticker/id.
    :type prices: pd.DataFrame
    :param log_returns: whether to compute using log returns
    :type log_returns: bool, defaults to False
    :return:(daily) returns
    :type:  pd.DataFrame
    
    """
    if log_returns:
        return np.log(1+ prices.pct_change()).dropna(how='all')

    else:
        return prices.pct_change().dropna(how='all')
    
    
def return_model(prices, method = "mean_historical_return", **kwargs):
    """
    Compute an estimate of future returns, using the return model specified in method.
    
    :param prices:adjusted closing prices of the asset, each row is a date and columns is a ticker/id.
    :type prices:pd.DataFrame
    :param returns_data: if true, the first argument is returns instead of prices.
    :type return_data: bool, defaults to False.
    :param method: the return model to use. should be one of:
        - "mean_historical_return"
        - "ema_historical_return"
        - "cpam_return"
    :type method : str, optional
    :raises NotImplementedError: if the supplied method is not recognised.
    :return: annualised smaple covariance matrix
    :type:pd.DataFrame
    
    """
    if method =="mean_historical_return":
        return mean_historical_return(prices, **kwargs)
    elif method =="ema_historical_return":
        return ema_historical_return(prices, **kwargs)
    elif method =="cpam_return":
        return capm_return(prices, **kwargs)
    else:
        raise NotImplementedError("Return model {} not implemented".format(method))
        

def mean_historical_return(prices, returns_data=False, compounding=True, frequency=250):
    """
    Calculate annulised mean (daily) historical return from input (daily) asset prices.
    Use 'compounding' to toggle between the default geometric mean (CAGR) and the arithmetic mean.
    
        
    :param prices:adjusted closing prices of the asset, each row is a date and columns is a ticker/id.
    :type prices:pd.DataFrame
    :param returns_data: if true, the first argument is returns instead of prices.
                    these **should not** be log returns.
                    
    :type returns_data: bool, defaults to False.
    :param compounding: compute geometric mean returns if True, arithmetic otherwise, optional.
    :type compounding:bool, defaults to True
    :param frequency: number of time periods in a year, defaults to 250(the number of trading days in a year)
    :type frequency: int, optioinal
    :return: annualised mean(daily) return for each asset
    :type:pd.Series
    """
    if not isinstance(prices, pd.DataFrame):
        warnings.warn("prices are not in a dataframe", RuntimeWarning)
        prices = pd.DataFrame(prices)
    if returns_data:
        returns = prices
    else:
        returns = returns_from_prices(prices)
        
    
    if compounding:
        return (1+returns).prod() **(frequency/returns.count()) -1
    else:
        return returns.mean() * frequency
    
    
def ema_historical_return(prices, returns_data=False, compounding=True, span=500,frequency=250):
    """
    Calculate the exponentially-weighted mean of (daily) historical returns, giving higher weight to more recent data.
    
    
        
    :param prices:adjusted closing prices of the asset, each row is a date and columns is a ticker/id.
    :type prices:pd.DataFrame
    :param returns_data: if true, the first argument is returns instead of prices.
                    these **should not** be log returns.
                    
    :type returns_data: bool, defaults to True.
    
    :param frequency: number of time periods in a year, defaults to 250(the number of trading days in a year)
    :type frequency: int, optioinal
    :param span:the-time-span for the EMA, defaults to 500-day EMA
    :type span:int, optional
    :return: annualised exponentially-weighted mean(daily) return of eeach asset
    :type:pd.Series
    
    """
    if not isinstance(prices, pd.DataFrame):
        warnings.warn("prices are not in a dataframe", RuntimeWarning)
        prices = pd.DataFrame(prices)
    if returns_data:
        returns = prices
    else:
        returns = returns_from_prices(prices)
    if compounding:
        return (1+ returns.ewm(span=span).mean().iloc[-1]) ** frequency -1
    else:
        return (returns.ewm(span=span).mean().iloc[-1]) ** frequency
    

    
    
    
def capm_return(prices, market_prices=None, returns_data=False, risk_free_rate=0.00, compounding=True,frequency=250):
    """
    Compute a return estimate using the Capital Asset Pricing Model. Under the CAPM, asset returns are equal to
    market returns plus a : math:'\beta' term encoding the relativerisk of the asset.

        
    :param prices:adjusted closing prices of the asset, each row is a date and columns is a ticker/id.
    :type prices:pd.DataFrame
    :param market_prices: adjusted cloing prices of the benchmark, defaults to None
    :type market_prices: pd.DataFrame, optioinal
    :param returns_data: if true, the first argument is returns instead of prices.
                    these **should not** be log returns.
                    
    :type returns_data: bool, defaults to True.
    
    :param frequency: number of time periods in a year, defaults to 250(the number of trading days in a year)
    :type frequency: int, optioinal
    :param span:the-time-span for the EMA, defaults to 500-day EMA
    :type span:int, optional
    :return: annualised exponentially-weighted mean(daily) return of eeach asset
    :type:pd.Series
    
    """
    if not isinstance(prices, pd.DataFrame):
        warnings.warn("prices are not in a dataframe", RuntimeWarning)
        prices = pd.DataFrame(prices)
    if returns_data:
        returns = prices
    else:
        returns = returns_from_prices(prices)
    if compounding:
        return (1+ returns.ewm(span=span).mean().iloc[-1]) ** frequency -1
    else:
        return (returns.ewm(span=span).mean().iloc[-1]) ** frequency
```
